```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#function that returns dy/dx
def model(y,x):
    dy_dx = np.cos(x) + (x*x)
    return dy_dx
```

```python
#initial condition
y0 = 0

#range of x
x = np.linspace(0, 10, 1000)

#solve ODE
y = odeint(model,y0,x)
```

```python
#plot results
plt.figure(figsize=[9,6])
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y(x)')
plt.grid()
plt.show()
```
