```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def apocalypse(y,t,P,B,n,G,A):
    
    S, Z, R = y
    
    dS_dt = P - B*S*Z - n*S
    dZ_dt = B*S*Z + G*R - A*S*Z
    dR_dt = n*S + A*S*Z - G*R
    
    return ([dS_dt, dZ_dt, dR_dt])
```

```python
#initial conditions
S0 = 500
Z0 = 0
R0 = 0

#parameter values
P = 0
B = 0.0095
n = 0.0001
G = 0.0001
A = 0.0001

#time range
t = np.linspace(0, 5, 1000)

#solve
solution = odeint(apocalypse, [S0, Z0, R0], t, args=(P,B,n,G,A))
solution = np.array(solution)
```

```python
#plot the graphs
plt.figure(figsize=[9,6])
plt.plot(t,solution[:, 0], label="S(t)")
plt.plot(t,solution[:, 1], label="Z(t)")
plt.plot(t,solution[:, 2], label="R(t)")
plt.grid()
plt.legend()
plt.xlabel("Time")
plt.ylabel("Number of people")
plt.title("Zombieland")
plt.show()
```
