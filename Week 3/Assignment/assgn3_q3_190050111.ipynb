```python
import numpy as np
from scipy.integrate import odeint
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def spring(y,t,m,g,k):
    
    X, V = y
    
    dX_dt = V
    dV_dt = g - ((k*X)/m)
    
    return ([dX_dt, dV_dt])
```

```python
#initial conditions
X0 = 0
V0 = 0

#parameter values
m = 1.5
g = 9.8
k = 2.5

#time range
t = np.linspace(0, 10, 10000)

solution = odeint(spring, [X0, V0], t, args=(m,g,k))
solution = np.array(solution)
```

```python
plt.figure(figsize=[9,6])
plt.plot(t,solution[:, 0], label="X(t)")
plt.plot(t,solution[:, 1], label="V(t)")
plt.grid()
plt.legend()
plt.xlabel("Time")
plt.ylabel("Position/velocity")
plt.title("Spring-mass system")
plt.show()
```
