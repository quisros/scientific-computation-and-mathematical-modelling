```python
#importing required modules
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import clear_output
%matplotlib inline
```

```python
# Set maximum number of iterations
maxIter = 500

# Set dimensions and delta
lenX = 201
lenY = 101
delta = 1

# Boundary conditions
Ttop = np.empty(lenX)
for i in range(lenX):
    Ttop[i] = 100 + 50*np.sin(2*np.pi*(i*delta/200))
Tbottom = 100
Tleft = 100
Tright = 100

# Initial guess of interior grid
Tguess = 100
```

```python
# Set colour interpolation and colour map
colorinterpolation = 100
colourMap = plt.cm.jet

# Set meshgrid
X, Y = np.meshgrid(np.arange(0, lenX), np.arange(0, lenY))
```

```python
# Set array size and set the interior value with Tguess
T = np.empty((lenY, lenX))
T.fill(Tguess)

# Set boundary conditions
T[(lenY-1):, :] = Ttop
T[:1, :] = Tbottom
T[:, (lenX-1):] = Tright
T[:, :1] = Tleft
```

```python
# Iteration (We assume that the iteration converges in maxIter = 500)
print("Iterating!")
for iteration in range(0, maxIter):
    for i in range(1, lenY-1, delta):
        for j in range(1, lenX-1, delta):
            T[i, j] = 0.1 * (T[i+1][j] + T[i-1][j] + 4*T[i][j+1] + 4*T[i][j-1])

print("Iteration finished.")
```

```python
# Configure the contour
plt.figure(figsize=[15, 7])
plt.style.use('dark_background')
plt.suptitle("Temperature Contours", fontsize=20)
plt.contourf(X, Y, T, colorinterpolation, cmap=colourMap)
plt.xlabel("X nodes", fontsize=15)
plt.ylabel("Y nodes", fontsize=15)
# Set Colorbar
plt.colorbar()
plt.show()
#PLEASE RUN THIS CELL TWICE TO GET PROPER LABELS ON THE PLOT
```
